```python
class Aluno:
    def __init__(self, nome: str, matricula: int, nota: float):
        self.nome = nome
        self.matricula = matricula
        self.nota = nota

    def __repr__(self):
        return f"Aluno(nome={self.nome}, matricula={self.matricula}, nota={self.nota})"
```

```python
class ColecaoGenerica:
    def __init__(self):
        self.elementos = []

    def criar_colecao(self):
        self.elementos = []

    def inserir(self, elemento):
        self.elementos.append(elemento)

    def remover(self, chave):
        for elemento in self.elementos:
            if (elemento.nome, elemento.matricula, elemento.nota) == chave:
                self.elementos.remove(elemento)
                return True
        return False

    def consultar(self, chave):
        for elemento in self.elementos:
            if (elemento.nome, elemento.matricula, elemento.nota) == chave:
                return elemento
        return None

    def listar(self):
        return self.elementos

    def destruir(self):
        if not self.elementos:
            del self.elementos
            return True
        return False

    def esvaziar(self):
        self.elementos = []
```

```python
# Criar a coleção
colecao = ColecaoGenerica()
colecao.criar_colecao()

# Inserir alunos
a1 = Aluno("Ana", 101, 8.5)
a2 = Aluno("Bruno", 102, 7.2)
a3 = Aluno("Carlos", 103, 9.0)

colecao.inserir(a1)
colecao.inserir(a2)
colecao.inserir(a3)

# Listar elementos
print("Coleção atual:")
for aluno in colecao.listar():
    print(aluno)

# Consultar
print("\nConsultando (Bruno, 102, 7.2):")
print(colecao.consultar(("Bruno", 102, 7.2)))

# Remover
print("\nRemovendo (Ana, 101, 8.5):")
colecao.remover(("Ana", 101, 8.5))

# Listar novamente
print("\nColeção após remoção:")
for aluno in colecao.listar():
    print(aluno)

# Esvaziar e destruir
colecao.esvaziar()
print("\nApós esvaziar:", colecao.listar())
print("Destruída:", colecao.destruir())
```
