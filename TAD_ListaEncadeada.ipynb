```python
# Definição do tipo escolhido
class Aluno:
    def __init__(self, nome: str, matricula: int, nota: float):
        self.nome = nome
        self.matricula = matricula
        self.nota = nota

    def __repr__(self):
        return f"Aluno(nome={self.nome}, matricula={self.matricula}, nota={self.nota})"
```

```python
# Definição do TAD Lista Encadeada
class No:
    def __init__(self, dado):
        self.dado = dado
        self.proximo = None

class ListaEncadeada:
    def __init__(self):
        self.cabeca = None

    def criar_colecao(self):
        self.cabeca = None

    def inserir(self, elemento):
        novo = No(elemento)
        novo.proximo = self.cabeca
        self.cabeca = novo

    def remover(self, chave):
        atual = self.cabeca
        anterior = None
        while atual:
            if (atual.dado.nome, atual.dado.matricula, atual.dado.nota) == chave:
                if anterior:
                    anterior.proximo = atual.proximo
                else:
                    self.cabeca = atual.proximo
                return True
            anterior = atual
            atual = atual.proximo
        return False

    def consultar(self, chave):
        atual = self.cabeca
        while atual:
            if (atual.dado.nome, atual.dado.matricula, atual.dado.nota) == chave:
                return atual.dado
            atual = atual.proximo
        return None

    def listar(self):
        elementos = []
        atual = self.cabeca
        while atual:
            elementos.append(atual.dado)
            atual = atual.proximo
        return elementos

    def destruir(self):
        if self.cabeca is None:
            del self.cabeca
            return True
        return False

    def esvaziar(self):
        self.cabeca = None
```

```python
# Programa de aplicação para testar o TAD Lista Encadeada

# Criar a coleção (lista)
lista = ListaEncadeada()
lista.criar_colecao()

# Inserir alunos
a1 = Aluno("Ana", 101, 8.5)
a2 = Aluno("Bruno", 102, 7.2)
a3 = Aluno("Carlos", 103, 9.0)

lista.inserir(a1)
lista.inserir(a2)
lista.inserir(a3)

# Listar elementos
print("Lista atual:")
for aluno in lista.listar():
    print(aluno)

# Consultar um aluno
print("\nConsulta por (Bruno, 102, 7.2):")
print(lista.consultar(("Bruno", 102, 7.2)))

# Remover um aluno
print("\nRemovendo (Ana, 101, 8.5):")
lista.remover(("Ana", 101, 8.5))

# Listar novamente
print("\nLista após remoção:")
for aluno in lista.listar():
    print(aluno)

# Esvaziar e destruir
lista.esvaziar()
print("\nApós esvaziar:", lista.listar())
print("Destruída:", lista.destruir())
```
