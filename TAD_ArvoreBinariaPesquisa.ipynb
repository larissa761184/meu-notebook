```python
class Aluno:
    def __init__(self, nome: str, matricula: int, nota: float):
        self.nome = nome
        self.matricula = matricula
        self.nota = nota

    def __repr__(self):
        return f"Aluno(nome={self.nome}, matricula={self.matricula}, nota={self.nota})"
```

```python
class NoArvore:
    def __init__(self, dado):
        self.dado = dado
        self.esquerda = None
        self.direita = None

class ArvoreBinariaPesquisa:
    def __init__(self):
        self.raiz = None

    def criar_colecao(self):
        self.raiz = None

    def inserir(self, elemento):
        if self.raiz is None:
            self.raiz = NoArvore(elemento)
        else:
            self._inserir_rec(self.raiz, elemento)

    def _inserir_rec(self, no, elemento):
        if elemento.matricula < no.dado.matricula:
            if no.esquerda is None:
                no.esquerda = NoArvore(elemento)
            else:
                self._inserir_rec(no.esquerda, elemento)
        else:
            if no.direita is None:
                no.direita = NoArvore(elemento)
            else:
                self._inserir_rec(no.direita, elemento)

    def consultar(self, chave):
        return self._consultar_rec(self.raiz, chave)

    def _consultar_rec(self, no, chave):
        if no is None:
            return None
        if (no.dado.nome, no.dado.matricula, no.dado.nota) == chave:
            return no.dado
        if chave[1] < no.dado.matricula:
            return self._consultar_rec(no.esquerda, chave)
        else:
            return self._consultar_rec(no.direita, chave)

    def remover(self, chave):
        self.raiz, removido = self._remover_rec(self.raiz, chave)
        return removido

    def _remover_rec(self, no, chave):
        if no is None:
            return no, False
        if chave[1] < no.dado.matricula:
            no.esquerda, removido = self._remover_rec(no.esquerda, chave)
        elif chave[1] > no.dado.matricula:
            no.direita, removido = self._remover_rec(no.direita, chave)
        else:
            if no.esquerda is None:
                return no.direita, True
            elif no.direita is None:
                return no.esquerda, True
            sucessor = self._minimo(no.direita)
            no.dado = sucessor.dado
            no.direita, _ = self._remover_rec(no.direita, (sucessor.dado.nome, sucessor.dado.matricula, sucessor.dado.nota))
            return no, True
        return no, removido

    def _minimo(self, no):
        atual = no
        while atual.esquerda:
            atual = atual.esquerda
        return atual

    def listar(self):
        elementos = []
        self._em_ordem(self.raiz, elementos)
        return elementos

    def _em_ordem(self, no, elementos):
        if no:
            self._em_ordem(no.esquerda, elementos)
            elementos.append(no.dado)
            self._em_ordem(no.direita, elementos)

    def destruir(self):
        if self.raiz is None:
            del self.raiz
            return True
        return False

    def esvaziar(self):
        self.raiz = None
```

```python
# Criar árvore
arvore = ArvoreBinariaPesquisa()
arvore.criar_colecao()

# Inserir alunos
a1 = Aluno("Ana", 101, 8.5)
a2 = Aluno("Bruno", 102, 7.2)
a3 = Aluno("Carlos", 103, 9.0)

arvore.inserir(a1)
arvore.inserir(a2)
arvore.inserir(a3)

# Listar em ordem
print("Elementos em ordem:")
for aluno in arvore.listar():
    print(aluno)

# Consultar
print("\nConsultando (Bruno, 102, 7.2):")
print(arvore.consultar(("Bruno", 102, 7.2)))

# Remover
print("\nRemovendo (Ana, 101, 8.5):")
arvore.remover(("Ana", 101, 8.5))

# Listar novamente
print("\nElementos após remoção:")
for aluno in arvore.listar():
    print(aluno)

# Esvaziar e destruir
arvore.esvaziar()
print("\nApós esvaziar:", arvore.listar())
print("Destruída:", arvore.destruir())
```
